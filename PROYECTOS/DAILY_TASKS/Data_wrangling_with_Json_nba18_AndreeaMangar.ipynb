```python
import requests
import json
import pandas as pd
```

### Data Mining & Data Wrangling
As you can see, there are jsons inside the original json. For that, we have to mining the json (data mining) to be able to use it correctly (data wrangling). 

How do you solve this issue? Research about this and try a solution. 

```python
r = requests.get(url='http://data.nba.net/prod/v2/2018/teams.json')
json_readed = json.loads(r.text)
#json_readed
```

```python
with open('data_NBA.json', 'w+') as outfile:
    json.dump(json_readed, outfile)
```

```python
with open('data_NBA.json', 'r+') as outfile:
    json_readed = json.load(outfile)
print(type(json_readed))
json_readed
```

```python
df_NBA = pd.DataFrame(json_readed)  # From dict
df_NBA
```

```python
type(df_NBA["league"])
```

```python
with open("data_NBA.json", "r+") as outfile:
    json_readed = json.load(outfile)
    NBA_json_list = json_readed["league"]["standard"]  # Real json
NBA_json_list
```

```python
with open('final_NBA_data.json', 'w+') as outfile:
    json.dump(NBA_json_list, outfile, indent=4) 
```

```python
df_NBA_json = pd.read_json("final_NBA_data.json")
df_NBA_json
```

```python
final_df = pd.concat([df_NBA, df_NBA_json],axis=1)
final_df
```

```python
#final_df = final_df.drop(["_internal"],axis=1)
final_df = final_df.drop(["_internal","league"],axis=1)
final_df = final_df.drop(['africa', 'eventName','pubDateTime','sacramento','standard','utah','vegas','xslt']) 
final_df
```
